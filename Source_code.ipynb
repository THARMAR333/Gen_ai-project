```python
import pandas as pd
```

```python
data = {
    "user": ["A", "B", "C", "D", "E"],
    "login_time": [9, 23, 14, 2, 18],
    "activity": ["normal", "download", "normal", "idle", "normal"]
}

df = pd.DataFrame(data)
print(df)
```

```python
def behavior_agent(row):
    if row["login_time"] < 6 or row["login_time"] > 22:
        return "⚠️ Anomaly"
    return "✅ Normal"
```

```python
def decision_agent(row):
    if row["login_time"] < 3:
        return "High Risk 🚨"
    elif row["login_time"] < 6 or row["login_time"] > 22:
        return "Medium Risk ⚠️"
    else:
        return "Low Risk ✅"
```

```python
def action_agent(row):
    if row["risk"] == "High Risk 🚨":
        return "🚫 Block User + Send Alert"
    elif row["risk"] == "Medium Risk ⚠️":
        return "📧 Send Warning"
    else:
        return "No Action"
```

```python
df["status"] = df.apply(behavior_agent, axis=1)
df["risk"] = df.apply(decision_agent, axis=1)
df["action"] = df.apply(action_agent, axis=1)

print(df)
```

```python
df.to_csv("output.csv", index=False)
```
