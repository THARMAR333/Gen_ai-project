```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

# load summarization model and tokenizer explicitly
# Using a common summarization model like 't5-small'
tokenizer = AutoTokenizer.from_pretrained("t5-small")
model = AutoModelForSeq2SeqLM.from_pretrained("t5-small")

text = """
Artificial Intelligence is one of the fastest growing technologies in the world.
It is used in healthcare, finance, education, transportation and many industries.
AI helps machines learn from data and make intelligent decisions.
"""

# Prepare input for the model
inputs = tokenizer(text, return_tensors="pt", max_length=512, truncation=True)

# Generate summary
# Parameters chosen to be similar to pipeline's max_length, min_length, and non-sampling (beam search instead of do_sample=False)
summary_ids = model.generate(
    inputs.input_ids,
    max_length=40,
    min_length=10,
    num_beams=4, # Use beam search for better quality, aligns with do_sample=False
    early_stopping=True # Stop when all beam hypotheses have finished
)
summary_text = tokenizer.decode(summary_ids[0], skip_special_tokens=True)


print("Original Text:")
print(text)

print("\nSummary:")
print(summary_text)
```
